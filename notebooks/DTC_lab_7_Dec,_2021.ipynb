```python
import matplotlib.pyplot as plt
import numpy as np
from math import pi
from scipy.signal import butter, lfilter
```

```python

Fc = 100
Fs = 10000
Ts = 1/Fs
m = 0.2
t = np.arange(0, 1, Ts)

Ac = 2
Fm = 2
carrier_signal = Ac*np.cos(2*pi*Fc*t)

Am = 1
msg_signal = Am*np.cos(2*pi*Fm*t)

am_signal = (Ac+Am*np.cos(2*pi*Fm*t))*np.cos(2*pi*Fc*t)


plt.subplot(3,1,1)
plt.title('Message signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.plot(t,msg_signal,'g')

plt.subplot(3,1,2)
plt.title('Carrier signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.plot(t,carrier_signal,'b')

plt.subplot(3,1,3)
plt.title('AM signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.plot(t,am_signal,'r')    

plt.rc('font',size=15)
fig=plt.gcf()
fig.set_size_inches(16,20)


dmod=2*carrier_signal**am_signal
nyq=0.5*Fs
normal_cutoff=(Fc/2)/nyq
```
